```python
import pandas as pd
import numpy as np
```

```python
# Re-reading the Excel file to get a fresh DataFrame 'df'
df = pd.read_excel('flight_price.xlsx')
```

```python
# Re-applying date feature engineering
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]

df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
# Re-applying arrival time feature engineering
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]

df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
# Re-applying departure time feature engineering
df['Departure_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_min'] = df['Dep_Time'].str.split(':').str[1]

df['Departure_hour'] = df['Departure_hour'].astype(int)
df['Departure_min'] = df['Departure_min'].astype(int)
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
# Handle 'Total_Stops' column:
# First, fill the single NaN value (as identified in previous info() and value_counts()).
# The previous intention was to map NaN to 1, representing '1 stop'.
df['Total_Stops'] = df['Total_Stops'].fillna('1 stop')

# Define the mapping for string values
stop_mapping = {'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4}

# Apply the mapping to the 'Total_Stops' column
df['Total_Stops'] = df['Total_Stops'].map(stop_mapping)

# Ensure the column is of integer type after mapping
df['Total_Stops'] = df['Total_Stops'].astype(int)

# Verify no NaN values remain and check unique values
print("Number of NaN values in 'Total_Stops' after processing:", df['Total_Stops'].isnull().sum())
print("Unique values in 'Total_Stops' after processing:", df['Total_Stops'].unique())
```

```python
# Drop the original 'Duration' column as we have now extracted hours and minutes
df.drop('Duration', axis=1, inplace=True)

# Display the first few rows of the DataFrame to show the changes
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
encoded_features = pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(), columns=encoder.get_feature_names_out())
```

```python
# Concatenate the encoded features with the original DataFrame
df = pd.concat([df, encoded_features], axis=1)

# Drop the original categorical columns as they are now encoded
df.drop(['Airline', 'Source', 'Destination'], axis=1, inplace=True)

# Display the first few rows of the updated DataFrame
df.head()
```
